## Importando o Tensor Flow

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

## Escolhendo o DataSet

```python
(x_train, y_train),(x_test, y_test) = tf.keras.datasets.cifar10.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

## InstГўnciando nosso Modelo de Rede Neuaral

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(10) 
])

model.summary()
```

## Compilar o Modelo

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

## Treinando o Modelo

```python
model.fit(x_train, y_train, epochs=30)
```

## Testando com imagens aleatГіrias

```python
test = model.predict(x_test)
```

```python
import numpy as np

cifar10_class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

print(f"PrediГ§Гөes para todas as {len(test)} imagens de teste:")
for i in range(len(test)):
    # Obter os logits para a i-Г©sima imagem de teste
    logits = test[i]
    
    # Encontrar o Гӯndice da classe com o maior logit (a classe predita)
    predicted_class_index = np.argmax(logits)
    
    # Obter o nome da classe usando o dicionГЎrio
    predicted_class_name = cifar10_class_names[predicted_class_index]
    
    print(f"Imagem de teste {i}: Classe predita = {predicted_class_name} (ГҚndice: {predicted_class_index})")
```

## Salvando a Rede Neural Treinada

```python
model.save('cifar10_model.h5')
```
